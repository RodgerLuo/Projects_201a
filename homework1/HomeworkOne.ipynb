```python
import csv
from pylab import *

salary = []
countries = []
years = []

# import salaries data
with open('Salaries.csv', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    for row in reader:
        salary.append(row);

#convert salary data from string to float
for i in range(0, len(salary)):
    for j in range(0, len(row)):
        salary[i][j] = float(salary[i][j])

# import data of countries and years
with open('CountriesAndYears.csv', newline='', encoding='utf-8') as g:
    readerCountries = csv.reader(g)  
    for row in readerCountries:
        countries.append(row[0])
        years.append(row[1])
        
figure(figsize(12,10))
imshow(salary, interpolation='nearest', vmin=0, vmax=50,  aspect='auto') # the aspect works with figsize
grid(True)
x = np.array(range(17))
y = np.array(range(34))
plt.yticks(y, countries)
plt.xticks(x, years)
xlabel('Data Source: U.S. Bureau of Labor Statistics, International Labor Comparisons, August 2013', fontsize = 8)
bar = colorbar()
bar.set_label('U.S. Dollars')
title('Hourly Direct Pay in Manufacturing, U.S. Dollars, 1996-2012')
```
