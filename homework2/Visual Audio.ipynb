```python
# this project is trying to apply time and frequency domain 
# of an audio clip into scatter plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fftpack import fft
framerate, soundfile = wavfile.read('summer.wav')

soundClip_ChannelOne = []
soundClip_ChannelTwo = []
colors = []
radius = []

#cut one piece out of the sound
start_sec = 20
end_sec = 20.1
start_sample = start_sec * framerate
end_sample = end_sec * framerate
soundClip = soundfile[start_sample:end_sample]

#seperate two channels
clip_length = len(soundClip)
for i in range(clip_length):
    soundClip_ChannelOne.append(int(soundClip[i][0]))
    soundClip_ChannelTwo.append(int(soundClip[i][1]))
    
N = clip_length
x = soundClip_ChannelOne
y = soundClip_ChannelTwo

#FFT to the soundClip
soundClip_fft = fft(soundClip)

#apply one channel to sound, and another to color
for i in range(len(soundClip_fft)):
    colors.append(abs(soundClip_fft[i][0]))
    radius.append(abs(soundClip_fft[i][1]))
radius = np.array(radius)

#normalise the radius
radiusNor = (radius - radius.min())/np.ptp(radius)

area = np.pi * (15 * radiusNor)**2
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
figure(figsize(10,10))
plt.show()

```
