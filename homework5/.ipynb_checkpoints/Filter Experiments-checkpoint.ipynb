```python
from scipy import misc
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from scipy.ndimage.filters import convolve as convolve2
import scipy.ndimage.morphology as morphology
import scipy.ndimage as ndimage
```

```python
land = imread('landscape.jpg')
figure(figsize=(8, 8))
imshow(land,aspect='auto')
land.shape
```

```python
landSum = sum(land.astype(float), axis=2)
figure(figsize=(8, 8))
imshow(landSum)
```

```python
kernel_stamp = zeros((40,40))
kernel_stamp[20,:] = 1
kernel_stamp[:,20] = 0.2
stamp = convolve2(landSum, kernel_stamp)
figure(figsize=(8, 8))
imshow(stamp, aspect='auto', interpolation='nearest')
```

```python
min_filter = ndimage.minimum_filter(stamp, (20,20))
figure(figsize=(8, 8))
imshow(min_filter)
```

```python
max_filter = ndimage.maximum_filter(stamp, (40,40))
figure(figsize=(8, 8))
imshow(max_filter)
```

```python
figure(figsize=(8, 8))
imshow((max_filter - min_filter))
```

```python
figure(figsize=(8, 8))
imshow((max_filter - min_filter)*(-1))
```

```python
figure(figsize=(8, 8))
reverse = max_filter - 2*min_filter
imshow(reverse)
```

```python
wtp = morphology.white_tophat(reverse, (4,1))
figure(figsize=(8, 8))
imshow(wtp)
```

```python
btp = morphology.black_tophat(reverse, (1,4))
figure(figsize=(8, 8))
imshow(btp);
```
