```python
img = imread('Mona_Lisa.png')
signal = img[0,:,0]
len(signal)
```

```python
win_start = arange(0, len(signal), 2)
print(win_start)
win_len = 2

mag_spectrum = []

for i in win_start:
    win = signal[i: i + win_len]
    X = fft.rfft(win)
    mag_spectrum.append(abs(X)/float(win_len/2))

plot(mag_spectrum)
```
