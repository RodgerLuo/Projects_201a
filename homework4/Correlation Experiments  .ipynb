```python
#The character locates on the fisrt row and first column. Let's dig it out.

whole = imread('1.png')
imshow(whole,aspect='auto')
whole.shape
```

```python
wholeSum = sum(whole.astype(float), axis=2)/3.0 #axis 2 to z axis
imshow(wholeSum)
colorbar()
```

```python
# Set the where parameter to 0.7 to make the character isolated while recognizable

wholeReady = where(wholeSum > 0.7, 1, 0) # return either 0 or 1
imshow(wholeReady)
colorbar()
```

```python
# Here is the pattern.

part = imread('2.png')
imshow(part)
colorbar()
part.shape
```

```python
partSum = sum(part.astype(float), axis=2)/3.0
imshow(partSum)
colorbar()
```

```python
#try different 'where' parameters: 0.6, 0.7, 0.8 
subplot(141)
partReady = where(partSum>0.6,1,0)
imshow(partReady,interpolation='nearest')
subplot(142)
partReady2 = where(partSum>0.7,1,0)
imshow(partReady2,interpolation='nearest')
subplot(143)
partReady3 = where(partSum>0.8,1,0)
imshow(partReady3,interpolation='nearest')
subplot(144)
imshow(part)
#It looks that 0.6 is the closest one to the original 
```

```python
# Let's do the cross-correlation
from scipy.signal import correlate2d
cc = correlate2d(wholeReady, partReady,mode='same')
imshow(cc)
colorbar()
#What?! It's recognized all over the place. 
#Let's try to make the sum of the pattern become zero (zero-padding? Not sure about the term), so it can reduce those errors.
```

```python
#making the sum of the pattern to zero
meanvalue=mean(partReady)
partUnity=partReady-meanvalue
```

```python
cc = correlate2d(wholeReady, partUnity,mode='same')
imshow(cc)
colorbar()
#Now, it's much better. We can see there is a red spot on the upper left.
```
